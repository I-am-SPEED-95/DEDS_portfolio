```python
#go_crm_train

import pandas as pd
import pyodbc

DB = {'servername': 'XPS15\\SQLEXPRESS', 'database': 'go_crm_train'}

export_conn = pyodbc.connect('DRIVER={SQL Server};SERVER=' + DB['servername'] + ';DATABASE=' + DB['database'] + ';Trusted_Connection=yes')

export_cursor = export_conn.cursor()
```

```python
#tabel naar dataframe
age_group_dataframe = pd.read_sql_query("SELECT * FROM age_group;", sqlite_conn)
country_dataframe = pd.read_sql_query("SELECT * FROM country;", sqlite_conn)
retailer_segment_dataframe =pd.read_sql_query("SELECT * FROM retailer_segment;", sqlite_conn)
sales_territory_dataframe = pd.read_sql_query("SELECT * FROM sales_territory;", sqlite_conn)
retailer_dataframe = pd.read_sql_query("SELECT * FROM retailer;", sqlite_conn)
retailer_site_dataframe = pd.read_sql_query("SELECT * FROM retailer_site;", sqlite_conn)
retailer_type_dataframe = pd.read_sql_query("SELECT * FROM retailer_type;", sqlite_conn)
sales_demographic_dataframe =pd.read_sql_query("SELECT * FROM sales_demographic;", sqlite_conn)
retailer_headquarters_dataframe =pd.read_sql_query("SELECT * FROM retailer_headquarters;", sqlite_conn)
retailer_contact_dataframe = pd.read_sql_query("SELECT * FROM retailer_contact;", sqlite_conn)
```

```python
#dataframe insert into tabel van database
#10 dataframes 1e dataframe age_group
# SQL INSERT statement
sql = "INSERT INTO age_group (AGE_GROUP_CODE, UPPER_AGE, LOWER_AGE) VALUES (?, ?, ?)"

# Data omzetten naar lijst van tuples
data_to_insert = age_group_dataframe[['AGE_GROUP_CODE', 'UPPER_AGE', 'LOWER_AGE']].values.tolist()

# Meerdere rijen tegelijk invoegen
try:
    export_cursor.executemany(sql, data_to_insert)
    export_conn.commit()
except Exception as e:
    print(" Fout bij executemany:", e)


# Opslaan en connectie sluiten
export_conn.commit()
export_cursor.close()
export_conn.close()
print("Data succesvol ingevoegd!")
```
