```python
import pandas as pd
import pyodbc
import sqlite3

DB = {'servername': 'XPS15\\SQLEXPRESS', 'database': 'gecombineerde_database'}

export_conn = pyodbc.connect('DRIVER={SQL Server};SERVER=' + DB['servername'] + ';DATABASE=' + DB['database'] + ';Trusted_Connection=yes')

export_cursor = export_conn.cursor()
```

```python
# SQLite bestand verbinden
sqlite_conn1 = sqlite3.connect('go_crm_train.sqlite')

# Haal de lijst met tabellen op
tables = pd.read_sql_query("SELECT name FROM sqlite_master WHERE type='table';", sqlite_conn1)

# Print de tabellen om te zien welke je wilt kopiÃ«ren
tables
```

```python
#tabel naar dataframe
country_dataframe = pd.read_sql_query("SELECT * FROM country;", sqlite_conn1)
country_dataframe
```

```python
#country tabel
# SQL INSERT statement
sql = "INSERT INTO country (COUNTRY_CODE, COUNTRY_EN, FLAG_IMAGE, SALES_TERRITORY_CODE) VALUES (?, ?, ?, ?)"

# Data omzetten naar lijst van tuples
data_to_insert = country_dataframe[['COUNTRY_CODE', 'COUNTRY_EN', 'FLAG_IMAGE', 'SALES_TERRITORY_CODE']].values.tolist()

# Meerdere rijen tegelijk invoegen
try:
    export_cursor.executemany(sql, data_to_insert)
    export_conn.commit()
except Exception as e:
    print(" Fout bij executemany:", e)


# Opslaan en connectie sluiten
export_conn.commit()
export_cursor.close()
export_conn.close()
print("Data succesvol ingevoegd!")
```
